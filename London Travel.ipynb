# Data Visualization

```python
#Importing needed libraries for visualization

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use("seaborn")
```

```python
df.hist(column='Visits (000s)', by='mode', bins=50, figsize=(12, 6))
```

```python
df.purpose.unique()
```

```python
freq =df.market.value_counts().head()
freq
```

```python
purpose =df.purpose.value_counts().head()
purpose
```

```python
purp = df.groupby('purpose')['Visits (000s)'].sum()
purp
```

```python
plt.figure(figsize=(20, 13))

plt.pie(purp, data=df, labels=['Business', 'Holiday', 'Miscellaneous', 'Study', 'VFR'], autopct="%1.1f%%",shadow=True)
plt.title("Distribution of Purpose")
plt.show()
```

#### Holiday came out as the top reason with 47.9%

```python
plt.figure(figsize=(12, 7))
purpose.plot(kind='bar')
plt.xticks(rotation = 0, weight='bold')
plt.title("Purposes by Rank")
for i, j in enumerate (purpose):
    plt.text(i, j, j, ha='center', weight='bold')
```

```python
plt.figure(figsize=(12, 7))
freq.plot(kind='bar')
plt.xticks(rotation = 0, weight='bold')
plt.title("Top Five Frequenting Countries")
for i, j in enumerate (freq):
    plt.text(i, j, j, ha='center', weight='bold')
```

#### France came top in visiting countries with a total of 2,611 visits

```python
mof  = df.groupby('mode')['Visits (000s)'].sum()
mof
```

```python
plt.figure(figsize=(20, 10))

color = ('gray','blue', 'brown')
plt.pie(mof, data=df, colors=(color), labels =('Air', 'Sea', 'Tunnel'), autopct="%1.1f%%", shadow=True)
plt.show()
```

#### Air happens to be the most used route bagging 79.8% of visits
